### Bar Plots

```python
# use pd to create a series of data in a table
new_d = pd.Series(np.random.rand(7), index=list('abcdefg'))
```

```python
new_d
```

```python
new_d.plot.bar(color='r', alpha=1) # Color: color of bars. alpha: visibility transparency. Can be adjusted
```

```python
# To show the bars horizontally we use the barh function as shown below
new_d.plot.barh(color='b', alpha=1)
```
