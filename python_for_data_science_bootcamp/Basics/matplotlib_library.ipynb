## Bar Graphs

```python
# How to show data in bar plots
new_films = ['film1', 'film2', 'film3', 'film4']
awards = [4, 6, 3, 7]
```

```python
plt.bar(range(len(new_films)), awards)
plt.title("Top Films")
plt.xlabel("Film")
plt.ylabel("no. of Awards")
plt.xticks(range(len(new_films)), new_films)
plt.show() # Show the bar graph
```

## Histograms

```python
# Histograms
# New data with 230 values, standard deviation around 15 and the values will concentrate around 150
new_h = np.random.normal(150, 15, 230) 
```

```python
plt.hist(new_h)
plt.show()
```

## Pie Charts

```python
x = np.array([37,27, 27, 13])
ch_labels = ["first", "second", "third", "fourth"]

plt.pie(x, labels = ch_labels)
plt.show()
```
